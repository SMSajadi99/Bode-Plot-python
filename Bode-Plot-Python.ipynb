```
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, bode

# Define the transfer function in MATLAB-style notation
numerator = [41.26 * 3541.74 * 16.83e-6, 41.26 + 3541.74 + 1]
denominator = [3541.74 * 16.83e-6, 1]

# Create the transfer function using scipy.signal.TransferFunction
sys = TransferFunction(numerator, denominator)

# Calculate the Bode frequency response using scipy.signal.bode
freq, mag, phase = bode(sys)

# Create a larger figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Plot the Bode magnitude (20*log10 magnitude) on the first subplot
ax1.semilogx(freq, mag))
ax1.set_ylabel('Magnitude (dB)', fontsize=14)
ax1.grid(which='both', axis='both', linestyle='--', linewidth=0.5)
ax1.set_title('Bode Diagram', fontsize=16)

# Plot the phase on the second subplot
ax2.semilogx(freq, phase)
ax2.set_xlabel('Frequency (rad/s)', fontsize=14)
ax2.set_ylabel('Phase (degrees)', fontsize=14)
ax2.grid(which='both', axis='both', linestyle='--', linewidth=0.5)

# Adjust tick label sizes for both subplots
ax1.tick_params(axis='both', labelsize=12)
ax2.tick_params(axis='both', labelsize=12)

plt.tight_layout()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, bode

# Define the transfer functions in MATLAB-style notation
numerator1 = [67.45 * 8245.55 * 8.578e-6, 67.45 + 8245.55 + 1]
denominator1 = [8245.55 * 8.578e-6, 1]

numerator2 = [41.26 * 3541.74 * 16.83e-6, 41.26 + 3541.74 + 1]
denominator2 = [3541.74 * 16.83e-6, 1]

numerator3 = [59.12 * 705.38 * 233.33e-6, 59.12 + 705.38 + 1]
denominator3 = [705.38 * 233.33e-6, 1]

# Create the transfer functions using scipy.signal.TransferFunction
sys1 = TransferFunction(numerator1, denominator1)
sys2 = TransferFunction(numerator2, denominator2)
sys3 = TransferFunction(numerator3, denominator3)

# Calculate the Bode frequency response using scipy.signal.bode
freq, mag1, phase1 = bode(sys1)
_, mag2, phase2 = bode(sys2)
_, mag3, phase3 = bode(sys3)

# Create a larger figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Plot the Bode magnitude (20*log10 magnitude) on the first subplot
ax1.semilogx(freq, mag1, label='a:  Unmodified PGE')
ax1.semilogx(freq, mag2, label='b: ETPGE')
ax1.semilogx(freq, mag3, label=' c: Zn(II)- MOFs/ETPGE')
ax1.set_ylabel('Magnitude (dB)', fontsize=14)
ax1.grid(which='both', axis='both', linestyle='--', linewidth=0.5)
ax1.set_title('Bode Diagram', fontsize=16)
ax1.legend(loc='upper right', fontsize=12)

# Plot the phase on the second subplot
ax2.semilogx(freq, phase1)
ax2.semilogx(freq, phase2)
ax2.semilogx(freq, phase3)
ax2.set_xlabel('Frequency (rad/s)', fontsize=14)
ax2.set_ylabel('Phase (degrees)', fontsize=14)
ax2.grid(which='both', axis='both', linestyle='--', linewidth=0.5)

# Adjust tick label sizes for both subplots
ax1.tick_params(axis='both', labelsize=12)
ax2.tick_params(axis='both', labelsize=12)

plt.tight_layout()
plt.show()
```
